```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    
    text = text.lower()
   
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    
    text = re.sub(r'http\S+', '', text)
   
    text = re.sub(r'[^a-z\s]', '', text)
  
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text


df = pd.read_csv('../data/raw/Tweets.csv')

# Temizleme işlemini uygula
df['clean_text'] = df['text'].apply(clean_text)

print("İlk temizlenmiş tweet örneği:")
print(df[['text', 'clean_text']].head())
```

```python
import os
# 3. Label Encoding (Duyguları sayısallaştırma)
le = LabelEncoder()
df['label'] = le.fit_transform(df['airline_sentiment'])

# 4. Split Data (Train - Test)
train_df, test_df = train_test_split(
    df[['clean_text', 'label', 'tweet_created']], 
    test_size=0.2, 
    random_state=42, 
    stratify=df['label']
)

# 5.Save for Colab and Dashboard


train_df.to_csv("../data/processed/train_cleaned.csv", index=False)
test_df.to_csv("../data/processed/test_cleaned.csv", index=False)
```
